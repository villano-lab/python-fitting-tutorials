# Impact of Initial Guesses

Our fits before were relatively accurate. Let's explore what happens if we change our initial guesses for our parameters.

```python
#Imports
import numpy as np
import lmfit as lf
import matplotlib.pyplot as plt
```

### Setup

Let's use a sine function this time instead of the quadratic.</br>
Otherwise, our setup will be more or less the same.

```python
#Function Setup - y = f(x) = sin(x)

#Fakedata
xdata = np.linspace(0,10,50)
ydata = np.zeros(len(xdata))
noise = np.random.normal(0,1,len(xdata))
for i,n in enumerate(xdata):
    ydata[i] = np.sin(n)+noise[i]
sigdata = np.ones(len(xdata))

#Function
def f(x,a,b,c):
    return a*np.sin(b*x+c)

#Points representing function - this is necessary to plot f(x) for our comparison.
ycurve = np.sin(xdata)
```

```python
### Try 1: Three-Parameter Fit

#Setup
weighdata = 1/sigdata
l_mod = lf.Model(f)
params = l_mod.make_params(a=1, b=0, c=0)

#Do fit
l_fit = l_mod.fit(ydata, params, x=xdata, weights=weighdata)

#Define Stuff
l_dict = l_fit.best_values
l_a = l_dict['a'] 
l_b = l_dict['b']
l_c = l_dict['c']

#Create array to plot
l_curve = l_a*np.sin(l_b*xdata+l_c)

#Plotting
fig = plt.figure(figsize=(9.0,8.0))

plt.errorbar(xdata,ydata,yerr=sigdata,fmt='bo',label='Data')
plt.plot(xdata,ycurve,label='Function')
plt.plot(xdata,l_curve,linestyle='--',label='LmFit')

plt.legend()
plt.show()
```

That's definitely not correct. This is basically a flat line! Take a look at the parameters; we should have changed our frequency *b* to 1. 
A frequency of zero will give us a horizontal line, and a frequency near zero will look linear over short ranges. 
Let's try again with better fitting parameters.

```python
### Try 1: Three-Parameter Fit

params = l_mod.make_params(a=1, b=1, c=0) #Fixed

#Do fit
l_fit = l_mod.fit(ydata, params, x=xdata, weights=weighdata)

#Define Stuff
l_dict = l_fit.best_values
l_a = l_dict['a'] 
l_b = l_dict['b']
l_c = l_dict['c']

#Create array to plot
l_curve = l_a*np.sin(l_b*xdata+l_c)

#Plotting
fig = plt.figure(figsize=(9.0,8.0))

plt.errorbar(xdata,ydata,yerr=sigdata,fmt='bo',label='Data')
plt.plot(xdata,ycurve,label='Function')
plt.plot(xdata,l_curve,linestyle='--',label='LmFit')

plt.legend()
plt.show()

l_fit
```

----

This is much better. There are clearly some issues with it, but we also have very noisy data. 
All of our values, when taking error into account, overlap with our known correct values (except for b, which comes fairly close at *b - &sigma; = 1.03*).

Clearly our initial parameters make a difference when it comes to fit quality. While it would be interesting to see what happens without initial guesses, the results are rather anticlimatic -- LmFit won't run without initial guesses, and SciPy assigns initial guesses of 1 to each parameter if they aren't given.
